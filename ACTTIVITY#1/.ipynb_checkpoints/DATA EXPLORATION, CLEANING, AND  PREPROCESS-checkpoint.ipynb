> #### Blockplot

```python
def _boxplot(data,col):
    sns.boxplot(x=data[col] ,fliersize=10,whis=50 ,dodge=False ,color="red" )
    
def _plot(df,cols):
    plt.figure(figsize=(16,4))
    #histrogram
    plt.subplot(1,2,1)
    plt.hist(df[cols],alpha = 1)
    #box
    plt.subplot(1,2,2)
    sns.boxplot(df[cols])
    plt.show()
```

```python
_plot(currect_data,'X')
```

```python
_plot(currect_data,"Y")
```

## Remove outliers 

```python
currect_data.describe()
```

```python
def _findQ (df, cols):
    Q1 = df[cols].quantile(0.25)
    Q3 = df[cols].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return Q1 ,Q3 ,IQR
```

```python
def _removeOutlier(df,cols):
    Q1 = df[cols].quantile(0.25)
    Q3 = df[cols].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df = df[~((df[cols] < (Q1 - 1.5 * IQR)) |(df[cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
    return df
```

```python
Q1 , Q3 ,IQR = _findQ(currect_data,"X")
print("Q1",Q1)
print("Q3",Q3)
print("IQR",IQR)
```

```python
Q1 , Q3 ,IQR = _findQ(currect_data,"Y")
print("Q1",Q1)
print("Q3",Q3)
print("IQR",IQR)
```

```python
cols = ['X','Y']
cleaned_data =_removeOutlier(currect_data,cols)
cleaned_data.reset_index(drop=True)
```
